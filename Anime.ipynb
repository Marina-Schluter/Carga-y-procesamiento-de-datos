### 1.15- Reparación de columnas

- La columna "rank" tiene un 10,5% de nulos, pero ofrece información sobre el ranking en una lista, no es algo que podamos deducir con ningún cálculo, no la cambiamos

- La columna "episodes" se podría reparar añadiendo en los campos nulos la media de tiempo correspondiente al tipo indicado en la columna "type". En este caso, un 3,57% de nulos no es relevante para hacer los cambios y lo descartamos

```python
#Obtención de la media de episodios según el tipo de anime

mean_episodes_by_type= df.groupby("type")["episodes"].mean()    #Si decidimos cambiar los nulos de la columna "episodes" agrupamos por la columna type y le pedimos que calcule la media de la columna epidosios en base a cada grupo de type
mean_episodes_by_type                                    
```

```python
 #Reparación de nulos en la columna episodes usando la media de episodios correspondiente a cada tipo de anime
 
for anime_type, mean_episodes in mean_episodes_by_type.items():     
     df.loc[(df["type"]==anime_type)&(df["episodes"].isnull()),"episodes"]=mean_episodes


#Usando "for....items()", le pedimos que revise cada clave-valor del diccionario "mean_episodes_by_type". La clave es el tipo de anime y el valor la media de episodiso 
#Esta primera línea también podría escribirse así: "mean_episodes=mean_episodes_by_type[anime_type]"
#Seleccionamos las filas del df donde el valor de la columna "type" es igual al de la iteración en marcha y la clave de la columna "episodes" es un nulo
#Usando df.loc· accedemos a las filas específicas para asignar el valor "mean_episodes" a la columna "episodes" de esas filas
```

```python
#Comprobación de cambios en los nulos de la columna episodes. Ha pasado de 3.57% a 0.032%

df.isnull().sum()/len(df)*100 
```

- El resto de columnas restantes las reparamos de forma general usando la media, la moda o la mediana según el tipo de dato
    - Media y Mediana para los int64 y los float64
    - Moda para los object

```python
def reparar_nulos(df):                                                   #Creamos función para la reparación
    for col in df.columns:                                               #Creamos bucle que itere sobre cada columna dentro de las columnas del df
        if df[col].dtype == 'int64' or df[col].dtype == 'float64':       #Si el type en la columa es un número entero o decimal, aplicamos lo siguiente:
            df[col] = df[col].fillna(df[col].mean())                     #El contenido de la columna pasa a rellenarse con la media de la columna 
        else:                                                            #Si no es un número, aplicamos lo siguiente
            df[col] = df[col].fillna(df[col].mode()[0])                  #El contenido de la columna para a rellenarse con el contenido de la moda de la columna 
    return df                                                            #La función debe devolver el df
 
df=reparar_nulos(df)                                                     #Se aplica la función al df
```
